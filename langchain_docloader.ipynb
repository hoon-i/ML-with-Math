```python
from langchain.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://linguistics.hufs.ac.kr/linguistics/m01_s01.do")

data = loader.load()
print(data[0].page_content)
```

```python
from langchain.document_loaders import UnstructuredURLLoader

urls = [
    "https://www.hufs.ac.kr/hufs/11403/subview.do?enc=Zm5jdDF8QEB8JTJGYmJzJTJGaHVmcyUyRjIxODclMkYxNTkwNTAlMkZhcnRjbFZpZXcuZG8lM0Y%3D",
    "https://www.hufs.ac.kr/hufs/11403/subview.do?enc=Zm5jdDF8QEB8JTJGYmJzJTJGaHVmcyUyRjIxODclMkYxNTkwNDUlMkZhcnRjbFZpZXcuZG8lM0Y%3D"
]

loader = UnstructuredURLLoader(urls=urls)

data = loader.load()
data
```

PDF Document Loader

```python
from langchain.document_loaders import PyPDFLoader

loader = PyPDFLoader("파일경로")
pages = loader.load_and_split()
```

```python
from langchain.document_loaders import Docx2txtLoader

loader = Docx2txtLoader('파일경로')
data = loader.load()
```

**CSV Document Loader**

```python
from langchain.document_loaders import CSVLoader

loader = CSVLoader(file_path='파일경로',
                   csv_args = {
                    'delimiter':",",
                    'quotechar':'"',
                    'fieldnames':['id','name','smiles','fpt']
                   }

)

data = loader.load()
```
